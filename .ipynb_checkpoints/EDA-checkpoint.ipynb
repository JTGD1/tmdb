### genres

```python
Counter(tr.genres).most_common(30)
```

```python
genre_list = []

for i in range(len(tr)):
    genres = tr.genres[i].split(';')
    for g in genres:
        if g not in genre_list:
            genre_list.append(g)
```

```python
X = np.zeros((len(tr), len(genre_list))).astype('int')
```

```python
df_genre = pd.DataFrame(X, columns=genre_list)
```

```python
for i in range(len(tr)):
    genres = tr.genres[i].split(';')
    for g in genres:
        df_genre[g][i] = 1
```

```python
df_genre = df_genre.drop('', axis = 1).head()
```

### production countries

```python
idx = tr.loc[:, ['title', 'original_language', 'production_countries']][tr.n_production_countries == 0].index
```

```python
A = tr[tr.n_production_countries == 0].imdb_id
len(A)
```

```python
def get_country(row):
    a = row[2:]
    pc = ia.get_movie(a)['production countries']
    return pc
```

```python
tr.production_countries[idx] = tr.imdb_id[idx].map(lambda row: get_country(row))
```
