```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
tr = pd.read_csv('data/train_2.csv')
te = pd.read_csv('data/test_2.csv')

y = tr.revenue_log
tr.drop('revenue_log', axis = 1, inplace = True)
```

```python
# combine the two data
cut = len(tr)
df = pd.concat([tr, te])
```

## quarter, month & weekofday

```python
plt.figure(figsize = (10, 10))
cols = ['year', 'quarter', 'month', 'weekofday']

for i in range(4):
    plt.subplot(4, 1, i+1)
    sns.boxplot(x = df[cols[i]], y = df.budget_log)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
cols = ['quarter', 'month']

for i in range(2):
    plt.subplot(2, 1, i+1)
    sns.distplot(df.budget_log / df[cols[i]])
plt.show()
```
